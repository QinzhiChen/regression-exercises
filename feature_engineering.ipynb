```python
def select_kbest(x, y, k=2):
    kbest = SelectKBest(f_regression, k=k)
    kbest.fit(x, y)
    mask = kbest.get_support()
    return x.columns[mask]
```

```python
def rfe(x, y, k=2):
    rf = RFE(LinearRegression(), n_features_to_select=k)
    rf.fit(x, y)
    mask = rf.get_support()
    return x.columns[mask]
```
